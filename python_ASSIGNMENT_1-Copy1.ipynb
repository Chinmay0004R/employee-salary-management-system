```python
# INPUT FROM USER

# Employee Names
employee_name_1 = input("Enter Employee Name 1 : ")
employee_name_2 = input("Enter Employee Name 2 : ")
employee_name_3 = input("Enter Employee Name 3 : ")
employee_name_4 = input("Enter Employee Name 4 : ")
employee_name_5 = input("Enter Employee Name 5 : ")

print("---------------------------------------------")


# BASE PAY INPUT
employee_base_pay_1 = float(input("Enter Base Pay for Employee 1 : "))
employee_base_pay_2 = float(input("Enter Base Pay for Employee 2 : "))
employee_base_pay_3 = float(input("Enter Base Pay for Employee 3 : "))
employee_base_pay_4 = float(input("Enter Base Pay for Employee 4 : "))
employee_base_pay_5 = float(input("Enter Base Pay for Employee 5 : "))

print("---------------------------------------------")


# HRA PERCENTAGE INPUT
employee_HRA_1 = float(input("Enter HRA Percentage for Employee 1 : "))
employee_HRA_2 = float(input("Enter HRA Percentage for Employee 2 : "))
employee_HRA_3 = float(input("Enter HRA Percentage for Employee 3 : "))
employee_HRA_4 = float(input("Enter HRA Percentage for Employee 4 : "))
employee_HRA_5 = float(input("Enter HRA Percentage for Employee 5 : "))

print("---------------------------------------------")


# DA PERCENTAGE INPUT
employee_DA_1 = float(input("Enter DA Percentage for Employee 1 : "))
employee_DA_2 = float(input("Enter DA Percentage for Employee 2 : "))
employee_DA_3 = float(input("Enter DA Percentage for Employee 3 : "))
employee_DA_4 = float(input("Enter DA Percentage for Employee 4 : "))
employee_DA_5 = float(input("Enter DA Percentage for Employee 5 : "))

print("---------------------------------------------")


# PF PERCENTAGE INPUT
employee_PF_1 = float(input("Enter PF Percentage for Employee 1 : "))
employee_PF_2 = float(input("Enter PF Percentage for Employee 2 : "))
employee_PF_3 = float(input("Enter PF Percentage for Employee 3 : "))
employee_PF_4 = float(input("Enter PF Percentage for Employee 4 : "))
employee_PF_5 = float(input("Enter PF Percentage for Employee 5 : "))
```

```python
# making in persentaages 
HRA_1 = employee_HRA_1 / 100 
HRA_2 = employee_HRA_2 / 100 
HRA_3 = employee_HRA_3 / 100 
HRA_4 = employee_HRA_4 / 100 
HRA_5 = employee_HRA_5 / 100 

DA_1 = employee_DA_1 / 100
DA_2 = employee_DA_2 / 100
DA_3 = employee_DA_3 / 100
DA_4 = employee_DA_4 / 100
DA_5 = employee_DA_5 / 100

PF_1 = employee_PF_1 / 100 
PF_2 = employee_PF_2 / 100
PF_3 = employee_PF_3 / 100
PF_4 = employee_PF_4 / 100
PF_5 = employee_PF_5 / 100
```

```python
#absulute value 
absolute_value_hra_1 =  employee_base_pay_1 * HRA_1
absolute_value_hra_2 =  employee_base_pay_2 * HRA_2 
absolute_value_hra_3 =  employee_base_pay_3 * HRA_3 
absolute_value_hra_4 =  employee_base_pay_4 * HRA_4 
absolute_value_hra_5 =  employee_base_pay_5 * HRA_5 
```

```python
# Absolute DA
absolute_value_DA_1 =  employee_base_pay_1 * DA_1
absolute_value_DA_2 =  employee_base_pay_2 * DA_2
absolute_value_DA_3 =  employee_base_pay_3 * DA_3
absolute_value_DA_4 =  employee_base_pay_4 * DA_4
absolute_value_DA_5 =  employee_base_pay_5 * DA_5
```

```python
# PF Deduction
absolute_value_PF_1 =  employee_base_pay_1 * PF_1
absolute_value_PF_2 =  employee_base_pay_2 * PF_2
absolute_value_PF_3 =  employee_base_pay_3 * PF_3
absolute_value_PF_4 =  employee_base_pay_4 * PF_4
absolute_value_PF_5 =  employee_base_pay_5 * PF_5
```

```python
# Total Gross
employee_total_gross_1 = employee_base_pay_1 + employee_HRA_1 + employee_DA_1
employee_total_gross_2 = employee_base_pay_2 + employee_HRA_2 + employee_DA_2
employee_total_gross_3 = employee_base_pay_3 + employee_HRA_3 + employee_DA_3
employee_total_gross_4 = employee_base_pay_4 + employee_HRA_4 + employee_DA_4
employee_total_gross_5 = employee_base_pay_5 + employee_HRA_5 + employee_DA_5
```

```python
# Net pay
employee_net_pay_1 = employee_total_gross_1 - absolute_value_PF_1
employee_net_pay_2 = employee_total_gross_2 - absolute_value_PF_2
employee_net_pay_3 = employee_total_gross_3 - absolute_value_PF_3
employee_net_pay_4 = employee_total_gross_4 - absolute_value_PF_4
employee_net_pay_5 = employee_total_gross_5 - absolute_value_PF_5
```

```python

print(f"Employee name is {employee_name_1} and your base pay is {employee_base_pay_1}")
print(f"Employee name is {employee_name_2} and your base pay is {employee_base_pay_2}")
print(f"Employee name is {employee_name_3} and your base pay is {employee_base_pay_3}")
print(f"Employee name is {employee_name_4} and your base pay is {employee_base_pay_4}")
print(f"Employee name is {employee_name_5} and your base pay is {employee_base_pay_5}")
# Printing Absolute HRA Values

print("\n--- Absolute HRA Values ---")
print(f"Employee 1 HRA Value : {absolute_value_hra_1}")
print(f"Employee 2 HRA Value : {absolute_value_hra_2}")
print(f"Employee 3 HRA Value : {absolute_value_hra_3}")
print(f"Employee 4 HRA Value : {absolute_value_hra_4}")
print(f"Employee 5 HRA Value : {absolute_value_hra_5}")



# Printing Absolute DA Values

print("\n--- Absolute DA Values ---")
print(f"Employee 1 DA Value : {absolute_value_DA_1}")
print(f"Employee 2 DA Value : {absolute_value_DA_2}")
print(f"Employee 3 DA Value : {absolute_value_DA_3}")
print(f"Employee 4 DA Value : {absolute_value_DA_4}")
print(f"Employee 5 DA Value : {absolute_value_DA_5}")



# Printing PF Deductions

print("\n--- PF Deductions ---")
print(f"Employee 1 PF : {absolute_value_PF_1}")
print(f"Employee 2 PF : {absolute_value_PF_2}")
print(f"Employee 3 PF : {absolute_value_PF_3}")
print(f"Employee 4 PF : {absolute_value_PF_4}")
print(f"Employee 5 PF : {absolute_value_PF_5}")



# Printing Total Gross Salary

print("\n--- Total Gross Salary ---")
print(f"Employee 1 Gross Salary : {employee_total_gross_1}")
print(f"Employee 2 Gross Salary : {employee_total_gross_2}")
print(f"Employee 3 Gross Salary : {employee_total_gross_3}")
print(f"Employee 4 Gross Salary : {employee_total_gross_4}")
print(f"Employee 5 Gross Salary : {employee_total_gross_5}")



# Printing Net Pay

print("\n--- Employee Net Pay ---")
print(f"Employee 1 Net Pay : {employee_net_pay_1}")
print(f"Employee 2 Net Pay : {employee_net_pay_2}")
print(f"Employee 3 Net Pay : {employee_net_pay_3}")
print(f"Employee 4 Net Pay : {employee_net_pay_4}")
print(f"Employee 5 Net Pay : {employee_net_pay_5}")

print("\ncomparision")
# Employee 1
is_high_earner_1 = employee_net_pay_1 > 80000
is_mid_range_1 = employee_net_pay_1 >= 40000 and employee_net_pay_1 <= 80000
is_low_earner_1 = employee_net_pay_1 < 40000

print(f"Employee 1 -> High Earner: {is_high_earner_1}, "
      f"Mid Range: {is_mid_range_1}, "
      f"Low Earner: {is_low_earner_1}")


# Employee 2
is_high_earner_2 = employee_net_pay_2 > 80000
is_mid_range_2 = employee_net_pay_2 >= 40000 and employee_net_pay_2 <= 80000
is_low_earner_2 = employee_net_pay_2 < 40000

print(f"Employee 2 -> High Earner: {is_high_earner_2}, "
      f"Mid Range: {is_mid_range_2}, "
      f"Low Earner: {is_low_earner_2}")


# Employee 3
is_high_earner_3 = employee_net_pay_3 > 80000
is_mid_range_3 = employee_net_pay_3 >= 40000 and employee_net_pay_3 <= 80000
is_low_earner_3 = employee_net_pay_3 < 40000

print(f"Employee 3 -> High Earner: {is_high_earner_3}, "
      f"Mid Range: {is_mid_range_3}, "
      f"Low Earner: {is_low_earner_3}")


# Employee 4
is_high_earner_4 = employee_net_pay_4 > 80000
is_mid_range_4 = employee_net_pay_4 >= 40000 and employee_net_pay_4 <= 80000
is_low_earner_4 = employee_net_pay_4 < 40000

print(f"Employee 4 -> High Earner: {is_high_earner_4}, "
      f"Mid Range: {is_mid_range_4}, "
      f"Low Earner: {is_low_earner_4}")


# Employee 5
is_high_earner_5 = employee_net_pay_5 > 80000
is_mid_range_5 = employee_net_pay_5 >= 40000 and employee_net_pay_5 <= 80000
is_low_earner_5 = employee_net_pay_5 < 40000

print(f"Employee 5 -> High Earner: {is_high_earner_5}, "
      f"Mid Range: {is_mid_range_5}, "
      f"Low Earner: {is_low_earner_5}")
```
